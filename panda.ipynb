**Appending and Resetting Index** : Filtered data frames can be easily combined together in the following example.
df2 and df3 dataframes consist of top 5 and last 5 rows respectively. 

```python
df2=df.head()
df3=df.tail()
#df1=df2.append(df3)
df1=pd.concat([df2,df3],axis=0)
df1
```

In the above code the df1 dataframe retains old index values, therefore we need to reset the index staring from 0. In order to remove old index,use drop=True parameter

```python
df1.reset_index(drop=True)
```

**ILOC function**: The function is used to access an individual row of the dataframe if the datatype of index is integer.For instance if we want to access 4th row the following command will accomplish the task

```python
df1.iloc[4]
```

**LOC function**: It is also used to access the data frame wherein the index is non integer. In df1 dataframe we can make state as index and implement the LOC function.

```python
df1=df1.set_index('State')
```

```python
df1.loc['West Bengal']
```

**Groupby Function** : It is used for grouping the rows based on the distinct values in the given column or columns in data frame. For instance if we want to group the dataframe df by the column name State, it is quite easy to complete the task.

```python
state_df=df.groupby('State')
```

```python
for i in state_df:
    print(i)
```
